### Сначала собрать единый long dataframe

```python
import pandas as pd

def attach_model(df, model_name):
    out = df.copy()
    out["model"] = model_name
    return out

wvs_all = pd.concat([
    attach_model(qwen_wvs_metrics, "Qwen"),
    attach_model(llama3_1_wvs_metrics, "Llama 3.1"),
    attach_model(llama3_3_wvs_metrics, "Llama 3.3"),
    attach_model(gptoss_wvs_metrics, "GPT-OSS")
], ignore_index=True)

hof_corr_all = pd.concat([
    attach_model(qwen_hof_corr, "Qwen"),
    attach_model(llama3_1_hof_corr, "Llama 3.1"),
    attach_model(llama3_3_hof_corr, "Llama 3.3"),
    attach_model(gptoss_hof_corr, "GPT-OSS")
], ignore_index=True)

hof_mae_all = pd.concat([
    attach_model(qwen_hof_mae, "Qwen"),
    attach_model(llama3_1_hof_mae, "Llama 3.1"),
    attach_model(llama3_3_hof_mae, "Llama 3.3"),
    attach_model(gptoss_hof_mae, "GPT-OSS")
], ignore_index=True)
```

### Лучший первый график: WVS soft alignment

```python
import matplotlib.pyplot as plt
import pandas as pd

plot_df = wvs_all.copy()

fig, ax = plt.subplots(figsize=(14, 6))

for model in plot_df["model"].unique():
    sub = plot_df[plot_df["model"] == model]
    pivot = sub.groupby("method")["soft_alignment"].mean().reindex(["fc","fr","fo","fu"])
    ax.plot(pivot.index, pivot.values, marker="o", linewidth=2, label=model)

ax.set_title("WVS Soft Alignment by Prompting Method")
ax.set_xlabel("Method")
ax.set_ylabel("Soft Alignment")
ax.set_ylim(0, 100)
ax.legend()
plt.grid(alpha=0.3)
plt.show()
```

### Country facets

```python
countries = ["Germany", "Bangladesh", "USA"]

fig, axes = plt.subplots(1, 3, figsize=(18, 5), sharey=True)

for ax, country in zip(axes, countries):
    sub = wvs_all[wvs_all["country"] == country]
    for model in sub["model"].unique():
        tmp = sub[sub["model"] == model]
        pivot = tmp.groupby("method")["soft_alignment"].mean().reindex(["fc","fr","fo","fu"])
        ax.plot(pivot.index, pivot.values, marker="o", linewidth=2, label=model)
    ax.set_title(country)
    ax.set_ylim(0, 100)
    ax.grid(alpha=0.3)

axes[0].set_ylabel("Soft Alignment")
axes[-1].legend(bbox_to_anchor=(1.05, 1))
plt.suptitle("WVS Soft Alignment Across Countries")
plt.tight_layout()
plt.show()
```

```python
wvs_summary = (
    wvs_all.groupby(["model", "method"])
    .agg(
        soft_alignment=("soft_alignment", "mean"),
        hard_alignment=("hard_alignment", "mean"),
        n=("n", "sum")
    )
    .reset_index()
)

hof_summary = (
    hof_corr_all.groupby(["model", "method"])
    .agg(
        spearman_r=("spearman_r", "mean")
    )
    .reset_index()
)

mae_col = [c for c in hof_mae_all.columns if c.lower() in ["mae", "mean_absolute_error"]][0]

mae_summary = (
    hof_mae_all.groupby(["model", "method"])
    .agg(
        mae=(mae_col, "mean")
    )
    .reset_index()
)

summary_table = (
    wvs_summary
    .merge(hof_summary, on=["model", "method"], how="left")
    .merge(mae_summary, on=["model", "method"], how="left")
)

summary_table = summary_table.sort_values(["model", "method"])
display(summary_table)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def attach_model(df, model_name):
    out = df.copy()
    out["model"] = model_name
    return out

# 1. Собираем общий raw dataframe
wvs_raw_all = pd.concat([
    attach_model(qwen_wvs, "Qwen"),
    attach_model(llama3_1_wvs, "Llama 3.1"),
    attach_model(llama3_3_wvs, "Llama 3.3"),
    attach_model(gptoss_wvs, "GPT-OSS")
], ignore_index=True)

# 2. Приводим score и gold к числам
wvs_raw_all["score"] = pd.to_numeric(wvs_raw_all["score"], errors="coerce")
wvs_raw_all["gold"] = pd.to_numeric(wvs_raw_all["gold"], errors="coerce")

# 3. Считаем ошибки относительно эталона
wvs_raw_all["abs_error"] = (wvs_raw_all["score"] - wvs_raw_all["gold"]).abs()
wvs_raw_all["bias"] = wvs_raw_all["score"] - wvs_raw_all["gold"]

# 4. Проверяем первые строки
display(
    wvs_raw_all[
        ["model", "question_id", "country", "method", "score", "gold", "abs_error", "bias"]
    ].head()
)

# 5. Heatmap: средняя абсолютная ошибка vs gold
err_heat = (
    wvs_raw_all.groupby(["model", "method"])["abs_error"]
    .mean()
    .reset_index()
    .pivot(index="model", columns="method", values="abs_error")
)

# чтобы порядок методов был логичный
err_heat = err_heat.reindex(columns=["fc", "fr", "fo", "fu"])

fig, ax = plt.subplots(figsize=(8, 5))
im = ax.imshow(err_heat.values)

ax.set_xticks(range(len(err_heat.columns)))
ax.set_xticklabels(err_heat.columns)
ax.set_yticks(range(len(err_heat.index)))
ax.set_yticklabels(err_heat.index)

for i in range(len(err_heat.index)):
    for j in range(len(err_heat.columns)):
        val = err_heat.iloc[i, j]
        txt = "NaN" if pd.isna(val) else f"{val:.2f}"
        ax.text(j, i, txt, ha="center", va="center")

ax.set_title("Average Absolute Error vs Gold")
ax.set_xlabel("Method")
ax.set_ylabel("Model")
plt.colorbar(im)
plt.show()
```

### MAE native vs English

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

wvs_lang = gptoss_wvs.copy()

wvs_lang["score"] = pd.to_numeric(wvs_lang["score"], errors="coerce")
wvs_lang["gold"] = pd.to_numeric(wvs_lang["gold"], errors="coerce")

wvs_lang["abs_error"] = (wvs_lang["score"] - wvs_lang["gold"]).abs()

lang_mae = (
    wvs_lang
    .groupby(["country", "method", "language_mode"], as_index=False)
    .agg(mae=("abs_error", "mean"))
)

lang_delta = (
    lang_mae
    .pivot_table(
        index=["country", "method"],
        columns="language_mode",
        values="mae"
    )
    .reset_index()
)

lang_delta["delta_mae"] = lang_delta["english"] - lang_delta["native"]
```

```python
heatmap_df = lang_delta.pivot(
    index="country",
    columns="method",
    values="delta_mae"
)

plt.figure(figsize=(8, 4))
plt.imshow(heatmap_df, aspect="auto")
plt.colorbar(label="English MAE - Native MAE")
plt.xticks(range(len(heatmap_df.columns)), heatmap_df.columns)
plt.yticks(range(len(heatmap_df.index)), heatmap_df.index)
plt.title("WVS Language Effect Heatmap")
plt.xlabel("Method")
plt.ylabel("Country")
plt.show()
```

 '>' 0 → native better \
 '<' 0 → english better 

```python
plot_df = lang_delta.sort_values("delta_mae")

plt.figure(figsize=(10, 5))
plt.barh(
    plot_df["country"] + " | " + plot_df["method"],
    plot_df["delta_mae"]
)
plt.axvline(0)
plt.xlabel("English MAE - Native MAE")
plt.title("Native vs English Performance Difference (WVS)")
plt.show()
```

### Hofstede — profile comparison by language


```python
def compute_hofstede_dimensions(df: pd.DataFrame) -> pd.DataFrame:
    tmp = df.copy()

    tmp["score"] = pd.to_numeric(tmp["score"], errors="coerce")

    dims = (
        tmp.groupby(
            ["country", "language_mode", "dimension"],
            as_index=False
        )
        .agg(estimated_score=("score", "mean"))
    )

    return dims
gptoss_hof_dims = compute_hofstede_dimensions(gptoss_hof)
```

```python
country_name = "Germany"

plot_df = gptoss_hof_dims[
    gptoss_hof_dims["country"] == country_name
].copy()

plt.figure(figsize=(8, 4))

for lang in ["native", "english"]:
    tmp = plot_df[plot_df["language_mode"] == lang]
    plt.plot(tmp["dimension"], tmp["estimated_score"], marker="o", label=lang)

plt.title(f"Hofstede Profile: {country_name}")
plt.xlabel("Dimension")
plt.ylabel("Estimated Score")
plt.legend()
plt.show()
```
